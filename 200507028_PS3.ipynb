## Step 1. Import the necessary libraries
## Step 2. Import the dataset from this address.
## Step 3. Assign it to a variable called users


```python
import pandas as pd
import numpy as np
import csv

users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep='|',lineterminator='\n')
users.head()
```

## Step 4. Discover what is the mean age per occupation


```python
users.groupby('occupation')['age'].mean()
```

## Step 5. Discover the Male ratio per occupation and sort it from the most to the least


```python
age = pd.pivot_table(users,index=['occupation'],  columns = ["gender"] ,values = 'age',   aggfunc='count').fillna(0)


age['perc_male'] = (age['M']/(age['F']+ age['M'])*100)

age.sort_values(['perc_male'])
```

## Step 6. For each occupation, calculate the minimum and maximum ages

```python
age2 = pd.pivot_table(users,index=['occupation'],  values = 'age',   aggfunc={'age' : [min,max]}).fillna(0)

age2
```

## Step 7. For each combination of occupation and sex, calculate the mean age 

```python
age3 = pd.pivot_table(users,index=['occupation'], columns= ['gender'], values = 'age',   aggfunc={'age' : np.mean}).fillna(0)

age3
```

## Step 8. For each occupation present the percentage of women and men

```python
age4 = pd.pivot_table(users,index=['occupation'],  columns = ["gender"] ,values = 'age',   aggfunc='count').fillna(0)


age4['perc_male'] = (age4['M']/(age4['F']+ age4['M'])*100)
age4['perc_female'] = (age4['F']/(age4['F']+ age4['M'])*100)



age4.iloc[:,[2,3]]
```

```python
data = pd.read_table("wind.txt",sep = '\s+',  index_col = False,lineterminator='\n',parse_dates = [[0,1,2]])

data
```

## Step 4. Year 2061? Do we really have data from this year? Create a function to fix it
## and apply it

```python
def year_change(x):
  y = x.year - 100 if x.year > 1989 else x.year
  return datetime.date(y, x.month, x.day)


data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(year_change)

data
```

## Step 5. Set the right dates as the index. Pay attention at the data type, it should be
## datetime64[ns].

```python
print(type(data['Yr_Mo_Dy']))
```

```python
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])
data = data.set_index('Yr_Mo_Dy')

data
```

```python
import sys
sys.prefix
from plotly.offline import init_notebook_mode, iplot, plot
import plotly as py
init_notebook_mode(connected=True)
import plotly.graph_objs as go
import matplotlib.pyplot as plt
import matplotlib.pyplot as pp # import matplolib for basic visuals
%matplotlib inline 

romans = pd.read_csv("roman-emperor-reigns.csv")

assassinated = ["Assassinated"]

romans_assassinated = romans[romans.Cause_of_Death.isin(assassinated)]

romans_assassinated = len(romans_assassinated)

romans_not = len(romans) - romans_assassinated

pp.pie([romans_assassinated,romans_not],labels = ["Assassinated","Not Assassinated"],autopct="%1.1f%%")
pp.title("Emperors and how they were killed")
```
